# README

## Bandana014 Forecasting Project

In this assignment we have three products, all of which share a similar seasonal demand pattern with heavy spikes around the holiday/peak season.  My methodology was to 

 1. impute missing values when the product was not active
 2. combine all three products into a single time series
 3. compute seasonality and trend with seasonal trend loess STL decomposition
 4. forecast for the upcoming year with ARIMA

## Questions

Whether or not products in the same subfamily all share seasonality effects would be the big concern here.

### (2) What other models or methodologies would you like to explore and why did you choose this approach?

#### Why I chose this approach:
Seasonality seems to be the main characteristic of the sales data.  I wanted to estimate this in as stable a way as possible.  Using only the single BANDANA014 product only gives two periods with which to estimate weekly seasonality, thus the need to combine products.

I chose STL to model seasonality.  I also tried classical moving average approach to model seasonality but decided to go with STL because it can provide seasonality for beginning and end of the data, whereas classical decomposition will have missing seasonality where it is not able to find a centered moving average.

#### Other methodologies to explore:
I need to do more tuning on the forecasting model.  Tune the order of ARIMA and compare mape between ETL and ARIMA.
If more products were available I would try hierarchical forecasting.

### (3) What challenges and complexities would your model need to handle that are not handled in this prototype homework solution

 1. Stockout detection: At various times, product sales drop to zero for a few weeks (E.G. Bandana014 around October 2020). Flieber would really like a demand forecast model rather than a sales forecast model.  Since demand is censored during stockouts, my forecast model likely under-estimates demand for these periods. 
 2. How to impute data when the product is inactive: I imputed missing data when a product was not being sold.  Need more automated ways to do this with more products.
 3. Model purchasing Events: We also see an apparent surge in purchases during the beginning of the global pandemic in summer 2020.  Need to account for this and other sudden events that affect purchasing in the forecasting models.
